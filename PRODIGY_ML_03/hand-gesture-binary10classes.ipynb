```python
import cv2
import numpy as np
import math
from cvzone.ClassificationModule import Classifier
from cvzone.HandTrackingModule import HandDetector

cap = cv2.VideoCapture(0)
detector = HandDetector(maxHands=1)

# Ensure the file paths are correct
try:
    classifier = Classifier("gesture_model.h5")
except UnicodeDecodeError as e:
    print(f"Error decoding file: {e}")
except Exception as e:
    print(f"Error loading classifier: {e}")

offset = 20
imgSize = 300

labels = ['call_me','fingers_crossed','up','okay','paper','rock','rock_on','scissor','peace','thumbs']

while True:
    success, img = cap.read()
    if not success:
        print("Failed to capture image from camera.")
        break

    imgOutput = img.copy()
    hands, img = detector.findHands(img)
    if hands:
        hand = hands[0]
        x, y, w, h = hand['bbox']

        imgWhite = np.ones((imgSize, imgSize, 3), np.uint8) * 255

        imgCrop = img[y-offset:y+h+offset, x-offset:x+w+offset]

        if imgCrop.size == 0:
            print("Warning: imgCrop is empty")
            continue

        aspectRatio = h / w

        if aspectRatio > 1:
            k = imgSize / h
            wCal = math.ceil(k * w)
            if wCal == 0:
                print("Warning: Invalid width calculation")
                continue

            try:
                imgResize = cv2.resize(imgCrop, (wCal, imgSize))
            except cv2.error as e:
                print(f"Error resizing image: {e}")
                continue

            wGap = math.ceil((imgSize - wCal) / 2)
            imgWhite[:, wGap:wCal+wGap] = imgResize

        else:
            k = imgSize / w
            hCal = math.ceil(k * h)
            if hCal == 0:
                print("Warning: Invalid height calculation")
                continue

            try:
                imgResize = cv2.resize(imgCrop, (imgSize, hCal))
            except cv2.error as e:
                print(f"Error resizing image: {e}")
                continue

            hGap = math.ceil((imgSize - hCal) / 2)
            imgWhite[hGap:hCal+hGap, :] = imgResize

        try:
            prediction, index = classifier.getPrediction(imgWhite, draw=False)
        except Exception as e:
            print(f"Error getting prediction: {e}")
            continue

        cv2.rectangle(imgOutput, (x-offset, y-offset-70), (x-offset+400, y-offset+60-50), (0, 255, 0), cv2.FILLED)
        cv2.putText(imgOutput, labels[index], (x, y-30), cv2.FONT_HERSHEY_COMPLEX, 2, (0, 0, 0), 2)
        cv2.rectangle(imgOutput, (x-offset, y-offset), (x+w+offset, y+h+offset), (0, 255, 0), 4)

        cv2.imshow('ImageCrop', imgCrop)
        cv2.imshow('ImageWhite', imgWhite)

    cv2.imshow('Image', imgOutput)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
